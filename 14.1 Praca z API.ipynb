API - Application Programming Interface<br>
REST - Representational State Transfer, definiuje, w jaki sposób klient i serwer komunikują się między sobą.<br>
REST web service - serwis internetowy oparty na architekturze REST<br>
REST API - dostęp do web service'u za pomocą URL.

Metody HTTP służą wykonywaniu operacji na zasobach czyli danych dostępnych w serwisie internetowym. 
Są następujące:<br>
GET - otrzymanie danych z zasobu<br>
POST - utworzenie nowego zasobu<br>
PUT - aktualizacja istniejącego zasobu<br>
PATCH - częściowa aktualizacja istniejącego zasobu<br>
DELETE - usunięcie zasobu<br>


Kiedy wykonujemy jakieś operacje otrzymujemy kod zwrotny:<br>
2XX - sukces<br>
3XX - przekierowanie<br>
4XX - błąd po stronie klienta (naszej)<br>
5XX - błąd po stronie serwera

Endpoints - miejsca (URL), któe dają dostęp do zasobów serwisu internetowego.

```python
import requests
api_url = "https://api-dbw.stat.gov.pl/api/1.1.0/area/area-area?lang=pl"
response = requests.get(api_url)
```

```python
#response.close() #zamyka połączenie z serwerem
#response.status_code #kod zwrotny
response.json()
```

```python
import pandas as pd

dane_json = response.json()
df = pd.DataFrame(dane_json)
df.columns = ['id', 'nazwa', 'id_element_nadrzedny', 'id_poziomu', 'nazwa_poziomu', 'zmienne_w_czasie']
df[df.nazwa_poziomu == "Dziedzina"]
#df[df.id == 16]
```
