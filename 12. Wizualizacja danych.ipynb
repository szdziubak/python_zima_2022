```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns 
import pandas as pd
```

```python
#Wykres liniowy

dane = np.linspace(start = 0, stop = 20, num = 50) #równo rozłożone 50 puntków pomiędzy 0 a 20
dane

#rysujemy wykres
plt.plot(dane, dane+10, label = 'linia') #oś x, oś y, nazwa linii = "linia"
plt.show() #wyświetl wykres
```

```python
#wiele wykresów na jednej osi
x = [1,3,4,6,9,17]
y = [3,7,9,-2,18, -29]
plt.plot(x, y, label = "Zestaw danych 1", color = "red")
x = [5,7,9,10,15]
y = [-3, -2, -1, 0, 10]
plt.plot(x, y, label = "Zestaw danych 2", color = "green")

#nazwy osi
plt.xlabel("Oś X")
plt.ylabel("Oś Y")

#tytuł wykresu
plt.title("Wykres 1. Wiele linii")

#legenda
plt.legend()

plt.show()
```

```python
#Wykres punktowy

x = [1,3,4,6,9,17]
y = [3,7,9,-2,18, -29]
plt.scatter(x, y, marker = "X", color = "red", alpha = 1) #scatter a nie plot
#marker - jaki znak
#alpha - przezroczystość - 1 brak przezroczystości
x = [5,7,9,10,15]
y = [-3, -2, -1, 0, 10]
plt.scatter(x, y, marker = "*", color = "green")

plt.legend()
plt.show()
```

```python
#Wykres kołowy


#dane
jezyki = ["Python","Java", "C++","JavScript"]
wyniki = [8,6,5,14]

#tworzenie wykresu
plt.pie(wyniki, #dane
        labels=jezyki, #labels
        colors=['red','green','blue','yellow'], #kolejne kolory
        startangle=90, #zaczynamy od 90 stopni
        shadow= True) #cień

plt.title("Języki programowania")

plt.show()
```

```python
#Wykres kolumnowy

#dane
jezyki = ["Python","Java", "C++","JavScript"]
wyniki = [8,6,5,14]

# Tworzenie wykresu
plt.bar(jezyki, wyniki, color='red')


plt.show()
```

```python
#Histogram

wiek_pracownika = [21,28,32,34,35,35,37,42,47,55]
podziały = [20,30,40,50,60] #binds

#tworzenie histogramu
plt.hist(wiek_pracownika, podziały, rwidth=0.7, color = "red") #rwidth - względna szerokość kolumny
```

```python
#Wykres bąbelkowy
import matplotlib.pyplot as plt
import numpy as np
# Wielkość wykresu
plt.figure(figsize=(8,5))

# Utworzenie danych
kraje = ['Katar','Luksemburg','Singapur','Brunea','Irlandia','Norwegia','Zjednoczone Emiraty Arabskie','Kuwejt']
populacja = [2781682, 604245,5757499,428963,4818690,5337962,9630959,4137312]
pkb_per_capita = [130475, 106705, 100345, 79530, 78785, 74356,69382, 67000]

#za duże bąbelki, musimy przeskalować
skalowane_pkb_per_capita  = np.divide(pkb_per_capita, 100)

# Rysowanie wykresu
plt.scatter(kraje, populacja, s=skalowane_pkb_per_capita, cmap="Blues", edgecolors="grey", alpha=0.5)

# znaczniki na osi X obrócone o 45 stopni
plt.xticks(rotation=45)

plt.show()
```

```python
import pandas as pd
#!pip install seaborn
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
# Ramka danych
df=pd.DataFrame({'x':[1,3,5,7,9,11],'y':[10,25,35,33,41,59]})

# Utworzeie lmplot - scatterplot + regresja + odchylenia
sns.lmplot(x='x', y='y', data=df)

plt.show()
```

```python
df=pd.DataFrame({'x':['A','B','C','D','E','F'],'y':[110,125,235,133,341,59]})

#Barplot - wykres słupkowy
sns.barplot(x='x', y='y', data=df)

plt.show()
```

```python
url = "https://raw.githubusercontent.com/PacktPublishing/Python-Data-Analysis-Third-Edition/master/Chapter05/HR_comma_sep.csv"
df=pd.read_csv(url)

# histogram + dystrybucja
sns.distplot(df.satisfaction_level)

plt.show()
```

```python
# boxplot
sns.boxplot(data=df[['satisfaction_level','last_evaluation']])

plt.show()
```

```python
#Dane wbudowane
data = sns.load_dataset("iris")
#data.head()

# Tworzenie macierzy korelacji
macierz_korelacji=data.corr()
#print(macierz_korelacji)

# heatmap - mapa ciepła - korelogram
sns.heatmap(macierz_korelacji, annot=True)

plt.show()
```

```python
#wykres gęstości

sns.kdeplot(df.satisfaction_level)

plt.show()
```

```python
# Wykresy typu violin
sns.violinplot(data=df[['satisfaction_level','last_evaluation']])

plt.show()
```

```python
# Wykresy zliczania
sns.countplot(x='salary', data=df)

plt.show()
```

```python
#Joinplot: dystrybucja + wielowymiarowe liczebności
sns.jointplot(x='satisfaction_level', y='last_evaluation', data=df, kind="kde")

# Show figure
plt.show()
```

```python
data = sns.load_dataset("iris")

# Pair Plot - histogramy + scatterplot
sns.pairplot(data)

plt.show()
```
