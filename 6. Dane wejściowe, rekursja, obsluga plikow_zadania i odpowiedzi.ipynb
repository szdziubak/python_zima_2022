Napisz funkcję zwracającą konkretny element ciągu fibonacciego za pomocą pętli, rekursji i rekursji ogonowej

```python
#przypomnienie - funkcja z pętlą wyglądała tak:
def ciag_fibonacciego(ilosc_elementow):
    x_0 = ''
    x_1 = 0
    x_2 = 1
    print(f"{x_1}\n{x_2}")
    for i in range(3, ilosc_elementow+1):
        x_0 = x_1
        x_1 = x_2
        x_2 = x_0 + x_1
        print(x_2)
#trzeba teraz przerobić tą funkcję aby zwracała konkretny element
def ciag_fibonacciego_wyraz(n):
    x_0 = ''
    x_1 = 0
    x_2 = 1
    if n == 1:
        return x_1
    elif n==2:
        return x_2
    else:    
        for i in range(2, n+1):
            x_0 = x_1
            x_1 = x_2
            x_2 = x_0 + x_1
    return x_2
```

```python
ciag_fibonacciego_wyraz(10)
```

```python
def fibonacci_rekurencyjnie(n):
    if n <= 1:
        return n 
    return fibonacci_rekurencyjnie(n-1)+fibonacci_rekurencyjnie(n-2)
```

```python
fibonacci_rekurencyjnie(10)
```

```python
def fibonacci_rekurencyjnie_ogon(n, a=0, b=1):
    if n == 0:
        return a
    if n == 1:
        return b
    return  fibonacci_rekurencyjnie_ogon(n-1, b, a+b) #nowe n to n-1, nowe a to b a nowe b to a+b
```

```python
fibonacci_rekurencyjnie_ogon(10)
```

__Zadanie 6.__ <br>
Opis sekwencji znajduje się w linii zaczynającej się od ">".<br>
Pozostałe linie to sekwencja.<br>
Puste linie mają długość 0 lub 1.<br>
Otwórz plik z dodaniem informacji, czy jest to opis, sekwencja czy pusta linia.<br>

```python
with open(sciezka+r"/6. Pliki/atp6-sequences.fasta", 'rt') as plik:
    for l in plik:
        # Opis
        if l.startswith('>'):
            print("opis: ", end='')
        # Sekwencja
        elif len(l) > 1:
            print("sekwencja: ", end='')
        # Pusta
        else:
            print("pusta: ", end='')
        print(f"{l}", end='')
```

Utwórz słowniki i dodaj do listy jak: dane = [{'Marka':'VOLKSWAGEN, 'Liczba':15471}, ...]

```python
dane = []
with open(sciezka+r"/6. Pliki/MARKI_2019_05.csv", 'r') as plik:
    naglowki = plik.readline().rstrip().split('|')
    naglowki = list(filter(None, naglowki))
    for l in plik:
        l = l.rstrip().split('|')
        l = list(filter(None, l))
        slownik = {}
        for k, v in zip(naglowki, l):
            slownik[k] = v
        print(slownik)
        dane.append(slownik)
print('-----------------------------------------')
print(dane)
```
